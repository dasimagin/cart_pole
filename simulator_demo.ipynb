```python
from common import Config, State, generate_pyplot_animation
from simulator import CartPoleSimulator

from IPython.display import HTML
```

```python
import matplotlib.pyplot as plt

config = Config()

initial_state = State(
    cart_position=0.0,
    cart_velocity=0.0,
    pole_angle=3.0,
    pole_angular_velocity=0.0
)

simulator = CartPoleSimulator()
simulator.reset_to(config, initial_state)

trajectory = []
timestamps = []

timestamp = 0.0
step = 0.05
for _ in range(100):
    trajectory.append(simulator.get_state())
    timestamps.append(timestamp)

    simulator.advance(step)
    timestamp += step
```

```python
anim = generate_pyplot_animation(config, trajectory, timestamps=timestamps)
HTML(anim.to_jshtml())
```
